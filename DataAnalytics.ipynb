```python
# Data Analytics
import numpy as np

my_list = [1,2,3]
my_array = np.array([1,2,3])

print(type(my_array))
print(type(my_list))

#Creating a numpy array from a list
print(np.array(my_list))

# or from a list of list
my_matrix = [[1,2,3],[4,5,6],[7,8,9]]
print(np.array(my_matrix))

#Creating array using built in function
#Return evenly spaced values within a given interval
#start, stop, step
print(np.arange(0,10))
print(np.arange(0,11,2))    # print in space of 2

#generating arrays of zeros and ones
print(np.zeros(3))
print(np.zeros((5,5)))
print(np.ones(3))
print(np.ones((3,3)))

#return evenly spaced numbers over a specified interval
# start , stop, number of elements (and not step )
#unlike numpy.arrange(), the stop value is included in the result
#the spacing between values is automatically determined based on the specified number of values(num)
print(np.linspace(0,10,3))
print(np.linspace(0,5,20))
print(np.linspace(0,5,21))


#Random number array
print(np.random.rand(2))
print(np.random.rand(5,5))


#From normal distribution
print(np.random.randn(2))
print(np.random.randn(5,5))

#Random integers from low( inclusive ) to high (exclusive)
print(np.random.randint(1,100))  # Single random integers between 1 and 100
print(np.random.randint(1,100,10))  # 10 random integers between 1 and 100

#Seeding for reproducable result
np.random.seed(77)
print(np.random.rand(4))    # 4 random numbers    #but for all machines will get the same random numbers

#sorting
arr = np.arange(25) # number from 0 to 24
ranarr = np.random.randint(0,50,10)    # 10 random numbers bwetween 0 and 50
print(arr)
print(ranarr)

#reshape : Return an array containing the same data with in new shape
print(arr.reshape(5,5))

#Max, min and their index position
print(ranarr.max())
print(ranarr.argmax())
print(ranarr.min())
print(ranarr.argmin())
```

```python
import numpy as np

#Creating sample array
arr = np.arange(0,11)
print(arr)

#bracket indexing and selection - similar to python list
#get value at an index
print(arr[8])

#Get values in a range
print(arr[1:8])

#Broadcasting
#Setting a value with index range (Broadcasting)
arr[0:5]=100
print(arr)

#Important notes on slices
slice_of_arr = arr[0:6]
print(slice_of_arr)

#Change slice
slice_of_arr[:]=99
print(slice_of_arr)
print(arr)
#Changes made are also there in our original array!
#Data is not copies, it is a view of the original array
#This avoid a memory problem

#To get a copy, we need to be explicit
arr_copy = arr.copy()
print(arr_copy)
```

```python
#Conditional selection
arr = np.arange(1,10)
print(arr)

#Basic Arithmetic
print(arr + arr)
print(arr - arr)
print(arr * arr)

#This will raise a Warining on division by Zero , but not an error!
# It just fills the spot with nan
print(arr/arr)

# Also a warning (but not an error) relating to infifnity
print(1/arr)

print(arr**3)

# Universal functions
print(np.sqrt(arr))
print(np.exp(arr))
print(np.sin(arr))
print(np.log(arr))

# summary statistics
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.var())
print(arr.std())
```
